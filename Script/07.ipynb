## 07-1 빠진 데이터를 찾아라! - 결측치 정제하기

### 결측치 찾기

#### 결측치 만들기

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'sex'   : ['M', 'F', np.nan, 'M', 'F'],
                   'score' : [5, 4, 3, 4, np.nan]})
df
```

```python
df['score'] + 1
```

#### 결측치 확인하기

```python
pd.isna(df)  # 결측치 확인
```

```python
pd.isna(df).sum()  # 결측치 빈도 확인
```

### 결측치 제거하기
#### 결측치 있는 행 제거하기

```python
df.dropna(subset = ['score'])  # score 결측치 제거
```

```python
df_nomiss = df.dropna(subset = ['score'])  # score 결측치 제거된 데이터 만들기
df_nomiss['score'] + 1                     # score로 연산
```

#### 여러 변수에 결측치 없는 데이터 추출하기

```python
df_nomiss = df.dropna(subset = ['score', 'sex'])  # score, sex 결측치 제거
df_nomiss
```

#### 결측치가 하나라도 있으면 제거하기

```python
df_nomiss2 = df.dropna()  # 모든 변수에 결측치 없는 데이터 추출
df_nomiss2
```

## 07-2 이상한 데이터를 찾아라! - 이상치 정제하기
### 이상치 제거하기 - 존재할 수 없는 값

```python
df = pd.DataFrame({'sex'   : [1, 2, 1, 3, 2, 1], 
                   'score' : [5, 4, 3, 4, 2, 6]})
df
```

#### 이상치 확인하기

```python
df['sex'].value_counts(sort = False)
```

```python
df['score'].value_counts(sort = False)
```

#### 결측 처리하기

```python
# sex가 3이면 NaN 부여
df['sex'] = np.where(df['sex'] == 3, np.nan, df['sex'])
df
```

```python
# score가 5보다 크면 NaN 부여
df['score'] = np.where(df['score'] > 5, np.nan, df['score'])
df
```

```python
df.dropna(subset = ['sex', 'score']) \
       .groupby('sex') \
       .agg(mean_score = ('score', 'mean'))
```

```python
# sex, score 결측치 제거
# sex별 분리
# score 평균 구하기
```

------------------------------------------------------------------------
#### (알아 두면 좋아요) `np.where()`는 문자와 `NaN`을 함께 반환할 수 없습니다.

```python
df = pd.DataFrame({'x1' : [1, 1, 2, 2]})
df['x2'] = np.where(df['x1'] == 1, 'a', np.nan)  # 조건에 맞을 때 문자 부여
```

```python
df
```

```python
df.isna()
```

```python
# 결측치로 만들 값에 문자 부여
df['x2'] = np.where(df['x1'] == 1, 'a', 'etc')
```

```python
# 'etc'를 NaN으로 바꾸기
df['x2'] = df['x2'].replace('etc', np.nan)  
```

```python
df
```

```python
df.isna()
```

### 이상치 제거하기 - 극단적인 값

#### 상자 그림으로 극단치 기준 정하기

##### 1. 상자 그림 살펴보기

```python
mpg = pd.read_csv('mpg.csv')
import seaborn as sns
sns.boxplot(data = mpg, y = 'hwy')
```

##### 2.극단치 기준값 구하기

**(1) 1사분위수, 3사분위수 구하기**

```python
pct25 = mpg['hwy'].quantile(.25)
pct25
```

```python
pct75 = mpg['hwy'].quantile(.75)
pct75
```

**(2) IQR 구하기**

```python
iqr = pct75 - pct25
iqr
```

**(3) 하한, 상한 구하기**

```python
pct25 - 1.5 * iqr  # 하한
```

```python
pct75 + 1.5 * iqr  # 상한
```

#### 3. 극단치를 결측 처리하기

```python
# 4.5 ~ 40.5 벗어나면 NaN 부여
mpg['hwy'] = np.where((mpg['hwy'] < 4.5) | (mpg['hwy'] > 40.5), np.nan, mpg['hwy'])

# 결측치 빈도 확인
mpg['hwy'].isna().sum()
```

#### 4. 결측치 제거하고 분석하기

```python
mpg.dropna(subset = ['hwy']) \
   .groupby('drv') \
   .agg(mean_hwy = ('hwy', 'mean'))
```

```python
# hwy 결측치 제거
# drv별 분리
# hwy 평균 구하기
```
